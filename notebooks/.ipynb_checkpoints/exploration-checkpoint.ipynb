```python
import librosa
import scipy
import matplotlib.pyplot as plt
import numpy as np

wave = x['wave']['s']
f_s = utils.extract.f_s

# non-default:
n_fft = utils.extract.n_fft
frame_length = utils.extract.n_fft
hop_length = utils.extract.hop_length

# # default:
# n_fft = 2048
# frame_length = 2048
# hop_length = 512

features_1 = []

# features extracted using librosa:

features_1.append(librosa.feature.spectral_flatness(wave, hop_length=hop_length, n_fft=n_fft))             # Spectral Flatness
features_1.append(librosa.feature.rms(wave, hop_length=hop_length, frame_length=frame_length))                           # Volume
features_1.append(librosa.feature.zero_crossing_rate(wave, hop_length=hop_length, frame_length=frame_length))            # Zero Crossing Rate
features_1.append(librosa.feature.spectral_centroid(wave, f_s, hop_length=hop_length, n_fft=n_fft))        # Spectral Centroind
features_1.append(librosa.feature.spectral_bandwidth(wave, f_s, hop_length=hop_length, n_fft=n_fft))       # Spectral Bandwidth
features_1.append(librosa.feature.spectral_contrast(wave, f_s, hop_length=hop_length, n_fft=n_fft))        # Spectral Contrast
features_1.append(librosa.feature.spectral_rolloff(wave, f_s, hop_length=hop_length, n_fft=n_fft))         # Spectral Rolloff
features_1.append(librosa.feature.mfcc(wave, f_s, hop_length=hop_length, n_fft=n_fft))                     # Mel-Frequency Cepstral Coefficients - MFCC 
features_1.append(librosa.feature.chroma_stft(wave, f_s, hop_length=hop_length, n_fft=n_fft))              # Spectrogram
# features_1.append(librosa.feature.chroma_cqt(wave, f_s, hop_length=hop_length, frame_length=frame_length))               # Constant-Q Chromagram
# features_1.append(librosa.feature.chroma_cens(wave, f_s, hop_length=hop_length, bins_per_octave=frame_length))              # Chroma Energy Normalized
# features_1.append(librosa.feature.tonnetz(wave, f_s, hop_length=hop_length))                  # Tonnetz
```

```python
for i, f in zip(range(len(features_1)), features_1):
    print(i, f.shape)
```

```python
wave_padded = np.pad(wave, int(frame_length//2), mode='reflect')
wave_windows = librosa.util.frame(wave_padded, frame_length=frame_length, hop_length=hop_length)
```

```python
def remove_outlayers(signal, std_thresh=3):
    
    to_filter = np.abs(signal)/np.std(signal) > std_thresh
    out = 1.*signal
    out[to_filter] = 0
    
    return out
```

```python
features_2 = []

# features extracted using scipy:

features_2.append(np.std(wave_windows, axis=0)) # spectral std
features_2.append(scipy.stats.skew(wave_windows, axis=0)) # spectral skewness
features_2.append(scipy.stats.kurtosis(wave_windows, axis=0)) # spectral kurtosis
features_2.append(np.stack([scipy.stats.entropy(np.abs(w)) for w in wave_windows.T])) # spectral entropy

for i in range(len(features_2)):
    features_2[i] = np.expand_dims(features_2[i], axis=0)
```

```python
for i, f in zip(range(len(features_2)), features_2):
    print(i, f.shape)
```
